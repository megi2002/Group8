# Introduction

For this research black market listings on the dark web will be investigated. Illegal trading and trafficking are booming on the dark web. In 2016, Aldridge and Décary‐Hétu (2016) estimated the monthly drug revenue on a single dark web marketplace to be close to six million Dollars. The obvious legal implications aside, missing out on any taxes concerning these transactions is a significant loss of income to the state. Data on the dark web is hard to come by without knowing the exact URL addresses, and even then, it is virtually impossible to track crypto transactions and the volume of illegal goods that have been sold. That makes any information on product availability and seller behaviour even more valuable. For that reason, this report aims to show a clear overview of the available products encountered on different dark web marketplaces and their sellers. The products found will be categorised depending on their legality and type. This investigation also supplies information on specific sellers and sites which can be investigated further based on the information provided. To that end, 3 sellers will be recommended to the police.  Lastly, this study also helps to educate students on criminal behaviour on the web.

# Legality of Products found


In total 92% of the products found were labelled illegal and 8% of the products found were labelled legal. Below a pie chart representing these amounts as percentages of the total amount of products is shown.

The legality of the products often directly relates to the category the product belongs to. This way, the 6 products from the category of prescription drugs have all been labelled legal. Besides that, an unlocked phone and the purchasing of Instagram followers were also deemed legal.

Nonetheless, we do have to keep in mind that even the topics regarded as legal in our analysis, can be argued on their legality. In theory, selling a phone is of course legal, however, the context matters a lot as the phone might be illegally acquired and sold for profit. By purchasing a phone of which you know that it is stolen or obtained illegally you are part of the crime and are ‘handling stolen goods’. Therefore it very much depends on the buyer whether they can argue that they were not aware of the phone being stolen.

# Conclusion and Reflection

Through scanning different black markets on the dark web we have found 100 different products that are for sale on the black market. With a total of 58 different products, drugs were by far the most represented category we found in our search. Almost all products found were considered illegal. 

Categorising the items on their legality was found to be quite difficult. Making it as black and white as saying that a product is either legal or illegal disregards the nuances that are part of arguing legality. The buying of prescribed drugs can be legal if you actually have the drugs prescribed. On the dark web, however, no one is going to check for you whether you actually have a subscription to the drug which means we cannot judge the legality of someone buying the drug without that information..

We were surprised to find little ‘shocking’ or sketchy items. We did not find anything in the category of weapons or other immediately threatening materials, although we did expect to find some of that in the first place. This might also be because we simply did not reach these ‘darker’ places on the dark web and only got to the more accessible links that we could find with our quick searches on more well-known websites such as https://darknetmarketslink.com  and https://darknetone.com/markets/.
